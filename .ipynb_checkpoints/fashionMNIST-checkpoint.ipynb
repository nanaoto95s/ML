# 分類問題（kerasの練習）

## Fashion MNISTを使って、写真の分類を行うモデルを訓練する
https://www.tensorflow.org/tutorials/keras/classification?hl=ja

```python
# tensorflowとtf.kerasのインポート
import tensorflow as tf
from tensorflow import keras
```

```python
# ライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt
```

```python
# fashion MNISTのデータをインポート
fashion_mnist = keras.datasets.fashion_mnist
(train_images,train_labels),(test_images,test_labels) = fashion_mnist.load_data()
```

```python
# データセットにはクラス名が含まれていないため、
#あとで画像を出力するときのためにクラス名を保存しておく
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

### データを前処理

```python
plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
# →ピクセルの値が0から255の数値
```

```python
# NNにデータを投げる前に、これらの値を0から１までの範囲にスケールする
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
# 訓練用データの初めの25枚をクラス名付きで表示
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
plt.show
```

### モデルの構築

```python
# 層の設定
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Flatten
model = Sequential([
    Flatten(input_shape = (28,28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
```

Flatten層では、28×28の２次元配列のデータを、28×28=784ピクセルの一次元配列に変換する。
画像の中に積まれているピクセルの行を取り崩し、横に並べるイメージ。

Dense層は、128個のノードがあるRelu関数層。
最後の層は、10個のノードがあるsoftmax層。

### モデルのコンパイル

```python
model.compile(optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])
```

### モデルの訓練

```python
# 訓練を開始するには、model.fitメソッドを呼び出す
model.fit(train_images, train_labels, epochs=5)
```

正解率89%

### 正解率の評価

```python
# テスト用データセットに対するモデルの性能を評価
# evaluate:入力データにおける損失値を計算する
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print('\nTest accuracy:', test_acc)
```

### 実際に予測する（分類予測）

```python
# predict:入力サンプルに対する予測値の出力
predictions = model.predict(test_images)
```

```python
# 最初の予測
predictions[0]
```

10の衣料品のそれぞれに該当するかの確信度を表す。

```python
# どのラベルがもっとも確信度が高いかを確認
np.argmax(predictions[0])
```
